```python
from CoolProp.CoolProp import PropsSI as CP
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import get_test_data

#Listas
registro_fluidos = []
COP_lista = []
lista_n = []
Wcomp_total_lista = []
cont = []

#Lista de coeficientes politrpicos
for n in range (110000, 210000, 10000):
    n = n / 100000
    cont.append(n)
    
#Lista de caractersticas para as curvas do grfico
lista_cor = ['r','g','b', 'c', 'm', 'k']
lista_ls = ['-', '-.', ':']
caracteristicas = []
for ls in lista_ls:
    for cor in lista_cor:
        caracteristicas.append((cor, ls))

#Dados do problema
Pfrig = 120 * 1000 #[W]
T6 = -80 + 273.15 #[K]
T3 = -8 + 273.15 #[K]
T7 = -12 + 273.15 #[K]
T9 = 35 + 273.15 #[K]
fluid_high_lista = ['R600a', 'R12', 'R152a', 'R717', 'R404A', 'R134a']
fluid_low_lista = ['R170', 'R1150', 'R41']
drop_suc = 12000 #[Pa]
drop_desc = 24000 #[Pa]
efet = 0.7 #efetividade do trocador intermedirio

#Loop nos dois ciclos - variao dos refrigerantes e do coeficiente politrpico
for fluid_low in fluid_low_lista:
    for fluid_high in fluid_high_lista:
        #Presso crtica do fluido superior
        P_crit_high = CP('PCRIT', fluid_high)
        #Ciclo inferior
        P6 = CP('P', 'T', T6, 'Q', 1, fluid_low)
        P3 = CP('P', 'T', T3, 'Q', 0, fluid_low)
        cp6 = CP('CP0MASS', 'T', T6, 'Q', 1, fluid_low)
        cp3 = CP('CP0MASS', 'T', T3, 'Q', 0, fluid_low)
        h6 = CP('H', 'T', T6, 'Q', 1, fluid_low)
        h3 = CP('H', 'T', T3, 'Q', 0, fluid_low)
        P1 = P6
        P4 = P3
        #Comparao entre os calores especficos nos pontos 3 e 6
        if cp6 < cp3:
            T1 = efet * (T3 - T6) + T6
            h1 = CP('H', 'T', T1, 'P', P1, fluid_low)
            ha = h1
            Pa = P1 - drop_suc
            Ta = CP('T', 'H', ha, 'P', Pa, fluid_low)
            Pb = Pa
            Tb = Ta + 4 #considerei um superaquecimento de 4C na admisso
            vb = 1 / CP('D', 'T', Tb, 'P', Pb, fluid_low)
            P2 = P3
            Pc = P2 + drop_desc
            h4 = h3 + h6 - h1
            T4 = CP('T', 'H', h4, 'P', P4, fluid_low)
            h5 = h4
        else:
            T4 = T3 - efet * (T3 - T6)
            h4 = CP('H', 'T', T4, 'P', P4, fluid_low)
            h1 = h3 + h6 - h4
            T1 = CP('T', 'H', h1, 'P', P1, fluid_low)
            ha = h1
            Pa = P1 - drop_suc
            Ta = CP('T', 'H', ha, 'P', Pa, fluid_low)
            Pb = Pa
            Tb = Ta + 4 #considerei um superaquecimento de 4C na admisso
            vb = 1 / CP('D', 'T', Tb, 'P', Pb, fluid_low)
            P2 = P3
            Pc = P2 + drop_desc
            h5 = h4
        m_flow_low = Pfrig / (h6 - h5)
        #Ciclo superior
        P7 = CP('P', 'T', T7, 'Q', 1, fluid_high)
        h7 = CP('H', 'T', T7, 'Q', 1, fluid_high)
        hw = h7
        Pw = P7 - drop_suc
        Tw = CP('T', 'H', hw, 'P', Pw, fluid_high)
        Px = Pw
        Tx = Tw + 4 #superaquecimento de 4C na admisso
        vx = 1 / CP('D', 'T', Tx, 'P', Px, fluid_high)
        P9 = CP('P', 'T', T9, 'Q', 0, fluid_high)
        P8 = P9
        Py = P8 + drop_desc
        h9 = CP('H', 'T', T9, 'Q', 0, fluid_high)
        h10 = h9
        par_fluido = (fluid_low, fluid_high)
        registro_fluidos.append(par_fluido)
        #Registro dos graus de subresfriamento e superaquecimento
        grau_subresf = T3 - T4
        grau_superaq = T1 - T6
        #print('O grau de subresfriamento ({}, {})  igual a {:0.2f}C.'.format(fluid_low,fluid_high,grau_subresf))
        #print('O grau de superaquecimento ({}, {})  igual a {:0.2f}C.'.format(fluid_low,fluid_high,grau_superaq))
        for n in cont:
            #Ciclo inferior
            vc = (Pb * vb ** n / Pc)**(1 / n)
            hc = CP('H', 'P', Pc, 'D', 1 / vc, fluid_low)
            h2 = hc
            #Trabalho especfico do compressor do ciclo inferior
            w_low = n / (n-1) * Pb * vb * ((Pc/Pb)**((n-1)/n) - 1)
            #Trabalho do compressor do ciclo inferior
            Wcomp_low = m_flow_low * w_low
            #Calor trocado no trocador central
            Qcond_low =  m_flow_low * (h2 - h3)
            #Ciclo superior
            Qevap_high = Qcond_low
            m_flow_high = Qevap_high / (h7 - h10)
            vy = (Px * vx ** n / Py)**(1 / n)
            #Verificao da presso crtica
            if P_crit_high > P8:
                lista_n.append(n)
                #Trabalho especfico do compressor do ciclo superior
                w_high = n / (n-1) * Px * vx * ((Py/Px)**((n-1)/n) - 1)
                #Trabalho do compressor do ciclo superior
                Wcomp_high = m_flow_high * w_high
                #Trabalho total do ciclo
                Wcomp_total = Wcomp_high + Wcomp_low
                Wcomp_total_lista.append(Wcomp_total / 1000)
                #COP
                COP = Pfrig / Wcomp_total
                COP_lista.append(COP)
            else:
                 print ('A presso P8 = {:0.2f} kPa (n = {}, fluido inf = {}, fluido sup = {})  maior do que a ' 
               'presso crtica do {} e, por isso, o ponto no foi includo no grfico'.format(P8 / 1000,n,fluid_low,
                                                                                               fluid_high, fluid_high))               



COP_lista_max = []
for n in range (0, len(COP_lista), len(cont)):
    COP_lista_max.append(COP_lista[n])

COP_lista_min = []
for n in range (len(cont)-1, len(COP_lista), len(cont)):
    COP_lista_min.append(COP_lista[n])
                    
if len(lista_n) != 0:
    plt.rc('xtick', labelsize = 15)
    plt.rc('ytick', labelsize = 15)
    plt.figure(figsize = [15, 15], dpi = 300)
    for i in range (0, len(Wcomp_total_lista), len(cont)):
        eixo_x = []
        eixo_y = []
        for j in range (i, i + len(cont), 1):
            eixo_x.append(lista_n[j])
            eixo_y.append(Wcomp_total_lista[j])
        plt.plot(eixo_x, eixo_y, color = caracteristicas[int(i/len(cont))][0], 
                 linestyle = caracteristicas [int(i/len(cont))][1],
                 label=registro_fluidos[int(i/len(cont))])
        plt.xlabel('Coeficiente politrpico', size = 15)
        plt.ylabel('Trabalho total [kW]', size = 15)
    plt.legend()

    plt.figure(figsize = [15, 15], dpi = 300)
    for i in range (0, len(COP_lista), len(cont)):
        eixo_x = []
        eixo_y = []
        for j in range (i, i + len(cont), 1):
            eixo_x.append(lista_n[j])
            eixo_y.append(COP_lista[j])
        plt.plot(eixo_x, eixo_y, color = caracteristicas[int(i/len(cont))][0], 
                 linestyle = caracteristicas [int(i/len(cont))][1], 
                 label=registro_fluidos[int(i/len(cont))])
        plt.xlabel('Coeficiente politrpico', size = 15)
        plt.ylabel('COP', size = 15)
    plt.legend(prop={'size' : 15})
    plt.savefig('cs2trocadorinfpoliref.eps', format='eps', bbox_inches = 'tight')
    plt.show()
else:
    print('Todos os pontos so invlidos')
```
